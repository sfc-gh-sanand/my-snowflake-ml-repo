# Environment Parity Test Notebook
This notebook tests if the GitHub Actions pipeline successfully injected environment variables.

```python
import snowflake.snowpark as snowpark

# Placeholders for the GitHub Pipeline
target_db = "{{database}}"
target_schema = "{{schema}}"
target_wh = "{{warehouse}}"

def main(session: snowpark.Session):
    # Apply the injected environment context
    session.use_database(target_db)
    session.use_schema(target_schema)
    session.use_warehouse(target_wh)
    
    print(f"Context set to: {target_db}.{target_schema} using {target_wh}")
    
    return {
        "STATUS": "SUCCESS",
        "DB": session.get_current_database(),
        "WH": session.get_current_warehouse()
    }
```

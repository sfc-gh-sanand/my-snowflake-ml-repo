```python
import snowflake.snowpark as snowpark
import sys

def main(session: snowpark.Session):
    # Capture arguments from sys.argv
    all_args = " ".join(sys.argv[1:]).split()
    
    try:
        target_db = all_args[0]
        target_schema = all_args[1]
        target_wh = all_args[2]
    except IndexError:
        target_db = session.get_current_database()
        target_schema = session.get_current_schema()
        target_wh = session.get_current_warehouse()

    # Set context
    session.use_database(target_db)
    session.use_schema(target_schema)
    session.use_warehouse(target_wh)
    
    return f"Successfully running in {target_db}.{target_schema}"
```
